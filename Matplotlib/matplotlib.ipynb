# A Picture is worth a thosand words.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel("days", fontsize=30)
plt.ylabel("temperature", fontsize=30)
plt.title("weather", fontsize=30)
plt.plot(x,y,'rD--')
```

# AXES, Labels, Grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("days",fontsize=20)
plt.ylabel("temperature",fontsize=20)
plt.title("weather")

plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

#plt.legend(loc="best")---> Default
plt.legend(loc="upper right", shadow=True)

plt.grid()
```

```python
plt.plot(days,max_t,label="Max",color="red")
plt.scatter(days,min_t,label="Min")
plt.scatter(days,avg_t,label="Avg",color="green")

plt.legend()
```

# BarChart

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue)
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.ylabel("Revenue")
plt.title("US tech stocks")
plt.bar(xpos,revenue,label="Stocks")

plt.legend()
```

# Multiple BarGraph

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.ylabel("Revenue in US$")
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,label="revenue",width=0.4)
plt.bar(xpos+0.2,profit,label="profit",width=0.4)

plt.legend()
```

# Horizontal BarGraph

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.yticks(ypos,company)
plt.xlabel("Revenue in US$")

plt.barh(ypos,revenue,label="Revenue")

plt.legend()
```

```python
plt.yticks(ypos,company)
plt.xlabel("Revenue in US$")

plt.barh(ypos-0.2,revenue,label="Revenue",height=0.4)
plt.barh(ypos+0.2,profit,label="Profit",height=0.4)

plt.legend()
```

# Mutiple data samples in a histogram

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],
        label=['men','women'])
plt.legend()
plt.grid()
```

```python
plt.ylabel("Sugar Level")
plt.xlabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],
        label=['men','women'], orientation="horizontal")
plt.legend()
plt.grid()
```

# PieChart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius=1.5)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.3f%%')
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.3f%%', shadow=True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct="%0.2f%%", explode=[0,0.1,0.1,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct="%0.2f%%", explode=[0,0.1,0.1,0,0],startangle=45)
plt.show()
```
