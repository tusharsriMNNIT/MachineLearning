```python
data = {'act':[1, 0, 0, 1, 0, 1, 0, 0, 1, 0, 1, 0],
        'pred': [1, 1, 0, 1, 0, 1, 1, 0, 1, 0, 0, 0]
        }
```

```python
da = pd.DataFrame(data,columns=['act','pred'])
da
```

```python
confusion_matrix = pd.crosstab(da['act'],da['pred'],rownames=['Actual'],colnames=['Predicted'],margins=True)
```

```python
import seaborn as sn
sn.heatmap(confusion_matrix, annot=True)
```

```python
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
df.set_index('day',inplace=True)
df
```

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no_event'
})
new_df
```

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method="bfill")
new_df
```

```python
new_df = df.interpolate(method="time")
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how="all")
new_df
```

```python
new_df = df.dropna(thresh=1)
new_df
```
