```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Cars90.csv')
df.head()
```

```python
df.isnull().sum(axis=0)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
selected_columns = ['Manufacturer', 'Price', 'MPG.city',
                    'DriveTrain', 'EngineSize', 'Horsepower', 
                    'Weight', 'Origin']
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 8))

plt.scatter(df['EngineSize'], 
            df['Horsepower'])

plt.title('Horsepower vs Engine Size')

plt.xlabel('Engine Size (litres)')
plt.ylabel('Horsepower (bhp)')

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.head()
```

```python
df['Origin'] = le.fit_transform(df['Origin'])
```

```python
df = df[selected_columns]
df.head()
```

```python
df.shape
```

```python
len(df['Manufacturer'].unique())
```

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

```python
X = df.drop('Price',axis=1)
X.shape
```

```python
Y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
df1 = pd.DataFrame({'test' : y_test, 'predicted' : y_pred})
df1.head()
```

```python
plt.figure(figsize=(12,6))

plt.plot(y_test.values, label='Actual')
plt.plot(y_pred, label='Predicted')

plt.title('Predicted vs original data', fontsize=15)

plt.ylabel('Price')
plt.legend()

plt.show()
```
